```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Scrapig the songs

song_results = soup.find_all('h3', class_ = 'c-title')
```

```python
# There is pattern to present the song name.  I created a loop to iterate ant retrieve only the songs names

songs_name = []

for a in range(6, 404, 4):
    song = song_results[a]
    songs_name.append(song.text.strip())

songs_name
```

```python
# Getting the artists is more complex because there is no regular pattern to allow us to iterate

artists_results = soup.find_all('span', class_='c-label', attrs={'u-font-size-20@tablet': False})
artist_names = [artist.get_text(strip=True) for artist in artists_results if artist.get_text(strip=True)]
artist_names
```

```python
# The result came with a lot of polution, so I created a function to clean it
def artists_names_new(lst):
    artists = []
    temp_index = []
    for i, artist in enumerate(lst):
        if any(c.isalpha() for c in artist):
            artists.append(artist)
            if i > 0:
                temp_index.append(lst[i - 1])

    return artists, temp_index
```

```python
# The results still need more cleaning
artists, temp_index = artists_names_new(artist_names)
full_df = pd.DataFrame({'Artists': artists, 'Temp_Index': temp_index})
duplicated_rows  = full_df.duplicated(subset=['Artists', 'Temp_Index'], keep='first')
filtered_df = full_df[~duplicated_rows]
filtered_df = filtered_df[filtered_df['Artists'] != 'NEW'] 
filtered_df = filtered_df[filtered_df['Artists'] != 'RE-\nENTRY'] 
filtered_df = filtered_df.drop('Temp_Index', axis =1)
```
