```python
import requests

response = requests.get('https://www.billboard.com/charts/hot-100/', verify=False)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
songs = soup.find_all('h3', class_ = 'c-title')
```

```python
songs = soup.find_all('h3', class_ = 'c-title')
title = []
for i in range(6, 406, 4):
    title.append(songs[i].text.strip())
print(title)
```

```python
import pandas as pd

df_title = pd.DataFrame({
    'Title': title
})
df_title
```

```python
songs2 = soup.find_all('span', class_ = 'c-label')
```

```python
artist = []
for i in range(4, 819):
    artist.append(songs2[i].text.strip())
print(artist)
```

```python
len(artist)
```

```python
def is_name(s):
    # Check if 's' ends with a letter and meets other criteria for being a name
    return s[-1].isalpha() and not all(char.isdigit() for char in s)

names_only = [item for item in artist if is_name(item)]
print(names_only)

```

```python
df = pd.DataFrame(names_only)
names_only
```

```python
df['Artist'] = df['Artist'].str.upper().str.strip()

df = df[~df['Artist'].isin(['NEW', 'RE-\nENTRY'])]

df.reset_index(drop=True, inplace=True)

print(df)
```
